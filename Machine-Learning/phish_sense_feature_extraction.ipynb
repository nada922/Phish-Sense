```python
# importing required packages for this section
from urllib.parse import urlparse,urlencode
import ipaddress
import re
```

## **3.1.1 Domain of the URL**

```python
# 1.Domain of the URL (Domain)
def getDomain(url):
    domain = urlparse(url).netloc
    if re.match(r"^www\.", domain):
        domain = re.sub(r"^www\.", "", domain)
    return domain
```

## **3.1.2. IP Address in the URL**

```python
# 2.Checks for IP address in URL (Have_IP)
def havingIP(url):
  try:
    ipaddress.ip_address(url)
    ip = 1
  except:
    ip = 0
  return ip
```

## **3.1.3. "@" Symbol in URL**

```python
# 3.Checks the presence of @ in URL (Have_At)
def haveAtSign(url):
  if "@" in url:
    at = 1
  else:
    at = 0
  return at
```

## **3.1.4. Length of URL**

```python
# 4.Finding the length of URL and categorizing (URL_Length)
def getLength(url):
  if len(url) < 54:
    length = 0
  else:
    length = 1
  return length
```

## **3.1.5. Depth of URL**

```python
# 5.Gives number of '/' in URL (URL_Depth)
def getDepth(url):
  s = urlparse(url).path.split('/')
  depth = 0
  for j in range(len(s)):
    if len(s[j]) != 0:
      depth = depth+1
  return depth
```

## **3.1.6. Redirection "//" in URL**

```python
# 6.Checking for redirection '//' in the url (Redirection)
def redirection(url):
  pos = url.rfind('//')
  if pos > 6:
    if pos > 7:
      return 1
    else:
      return 0
  else:
    return 0
```

## **3.1.7. "http/https" in Domain name**

```python
# 7.Existence of “HTTPS” Token in the Domain Part of the URL (https_Domain)
def httpDomain(url):
  domain = urlparse(url).netloc
  if 'https' in domain:
    return 1
  else:
    return 0
```

```python
#listing shortening services
shortening_services = r"bit\.ly|goo\.gl|shorte\.st|go2l\.ink|x\.co|ow\.ly|t\.co|tinyurl|tr\.im|is\.gd|cli\.gs|" \
                      r"yfrog\.com|migre\.me|ff\.im|tiny\.cc|url4\.eu|twit\.ac|su\.pr|twurl\.nl|snipurl\.com|" \
                      r"short\.to|BudURL\.com|ping\.fm|post\.ly|Just\.as|bkite\.com|snipr\.com|fic\.kr|loopt\.us|" \
                      r"doiop\.com|short\.ie|kl\.am|wp\.me|rubyurl\.com|om\.ly|to\.ly|bit\.do|t\.co|lnkd\.in|db\.tt|" \
                      r"qr\.ae|adf\.ly|goo\.gl|bitly\.com|cur\.lv|tinyurl\.com|ow\.ly|bit\.ly|ity\.im|q\.gs|is\.gd|" \
                      r"po\.st|bc\.vc|twitthis\.com|u\.to|j\.mp|buzurl\.com|cutt\.us|u\.bb|yourls\.org|x\.co|" \
                      r"prettylinkpro\.com|scrnch\.me|filoops\.info|vzturl\.com|qr\.net|1url\.com|tweez\.me|v\.gd|" \
                      r"tr\.im|link\.zip\.net"
```

```python
# 8. Checking for Shortening Services in URL (Tiny_URL)
def tinyURL(url):
    match=re.search(shortening_services,url)
    if match:
        return 1
    else:
        return 0
```

## **3.1.9. Prefix or Suffix "-" in Domain**

```python
# 9.Checking for Prefix or Suffix Separated by (-) in the Domain (Prefix/Suffix)
def prefixSuffix(url):
    if '-' in urlparse(url).netloc:
        return 1            # phishing
    else:
        return 0            # legitimate
```

## **3.2.2. Web Traffic**

```python
# 12.Web traffic (Web_Traffic)
def web_traffic(url):
    try:
        # SimilarWeb and other APIs often block scripts.
        # We add a User-Agent header to look like a real browser.
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
        }
        # Encode the URL
        url_encoded = urllib.parse.quote(url)
        # Note: The URL you used in the traceback might be outdated or require an API key.
        # If this API returns 404, we catch it here.
        rank_url = "https://api.similarweb.com/batch/v4/request-report" + url_encoded

        response = requests.get(rank_url, headers=headers, timeout=5)
        if response.status_code != 200:
            return 1 # Default value if the site isn't ranked or API fails

        # Parse logic here based on SimilarWeb's actual XML/JSON response
        # (Assuming your original .find("REACH") logic worked)
        return 0
    except:
        return 1 # If any network error occurs, treat as "no traffic/unknown"
```

## **3.2.3. Age of Domain**

```python
# 13.Survival time of domain: The difference between termination time and creation time (Domain_Age)
def domainAge(domain_name):
  creation_date = domain_name.creation_date
  expiration_date = domain_name.expiration_date
  if (isinstance(creation_date,str) or isinstance(expiration_date,str)):
    try:
      creation_date = datetime.strptime(creation_date,'%Y-%m-%d')
      expiration_date = datetime.strptime(expiration_date,"%Y-%m-%d")
    except:
      return 1
  if ((expiration_date is None) or (creation_date is None)):
      return 1
  elif ((type(expiration_date) is list) or (type(creation_date) is list)):
      return 1
  else:
    ageofdomain = abs((expiration_date - creation_date).days)
    if ((ageofdomain/30) < 6):
      age = 1
    else:
      age = 0
  return age
```

## **3.2.4. End Period of Domain**

```python
# 14.End time of domain: The difference between termination time and current time (Domain_End)
from datetime import datetime

def domainEnd(domain_name):
    # Initialize the variable to None at the very start
    expiration_date = None

    try:
        # Perform the whois lookup
        w = whois.whois(domain_name)
        expiration_date = w.expiration_date

        # WHOIS sometimes returns a list of dates; take the first one
        if isinstance(expiration_date, list):
            expiration_date = expiration_date[0]

    except Exception:
        # If whois fails, expiration_date stays None
        pass

    # Now this check will work because the variable exists
    if not expiration_date:
        return 1
    else:
        today = datetime.now()

        # Clean the timezone (Fix for your previous TypeError)
        if expiration_date.tzinfo is not None:
            expiration_date = expiration_date.replace(tzinfo=None)

        try:
            end = abs((expiration_date - today).days)
            # If remaining time is less than 6 months (approx 180 days)
            return 1 if (end / 30) < 6 else 0
        except:
            return 1
```

## **3.3.1. IFrame Redirection**

```python
# 15. IFrame Redirection (iFrame)
def iframe(html_content):
  if html_content == "":
    return 1
  else:
    if re.findall(r"[<iframe>|<frameBorder>]", html_content):
      return 0
    else:
      return 1
```

## **3.3.2. Status Bar Customization**

```python
# 16.Checks the effect of mouse over on status bar (Mouse_Over)
def mouseOver(html_content):
    if html_content == "" :
      return 1
    else:
      if re.findall("<script>.+onmouseover.+</script>", html_content):
        return 1
      else:
        return 0
```

## **3.3.3. Disabling Right Click**

```python
# 17.Checks the status of the right click attribute (Right_Click)
def rightClick(html_content):
  if html_content == "":
    return 1
  else:
# Notice we just use html_content here, NOT html_content.text
    if re.findall(r"event.button ?== ?2", html_content):
      return 0
    else:
      return 1
```

## **3.3.4. Website Forwarding**

```python
# 18.Checks the number of forwardings (Web_Forwards)
def forwarding(response):
  if response == "" or response is None:
    return 1
  else:
    if len(response.history) <= 2:
      return 0
    else:
      return 1
```

## **4. Computing URL Features**
Create a list and a function that calls the other functions and stores all the features of the URL in the list. We will extract the features of each URL and append to this list.

```python
#Function to extract features
def featureExtraction(url, label, response):
    features = []

    # 1. Handle HTML content safely
    # If the loop failed to get a response, we use an empty string
    if response is not None:
        html_content = response.text
    else:
        html_content = ""

    # 2. Address Bar Based Features
    features.append(getDomain(url))
    features.append(havingIP(url))
    features.append(haveAtSign(url))
    features.append(getLength(url))
    features.append(getDepth(url))
    features.append(redirection(url))
    features.append(httpDomain(url))
    features.append(tinyURL(url))
    features.append(prefixSuffix(url))

    # 3. Domain Based Features (DNS/WHOIS)
    dns = 0
    domain_name = None
    try:
        domain_name = whois.whois(urlparse(url).netloc)
    except:
        dns = 1

    features.append(dns)
    features.append(web_traffic(url))
    features.append(1 if dns == 1 else domainAge(domain_name))
    features.append(1 if dns == 1 else domainEnd(domain_name))

    # 4. HTML & Javascript Features
    # These functions must take the STRING 'html_content'
    features.append(iframe(html_content))
    features.append(mouseOver(html_content))
    features.append(rightClick(html_content))
    features.append(forwarding(response))

    features.append(label)
    return features
```

```python

#Extracting the feautres & storing them in a list
import requests
import csv
import os

legi_features = []
label = 0
start_index = 4500 # If you crash at 3500, change this to 3500 and run again!
total_urls = 6100
checkpoint_interval = 100

# Use a session to manage network resources efficiently
session = requests.Session()

# Create or prepare the CSV file for saving progress
csv_filename = "legi_features_backup(4).csv"

for i in range(start_index, total_urls):
    url = legiurl['URLs'][i]

    # Wrap the ENTIRE process for this URL in a try block
    try:
        response = session.get(url, timeout=5)
        features = featureExtraction(url, label, response)
    except Exception as e:
        # If ANYTHING goes wrong (network error, extraction error), catch it
        print(f"Error processing URL {i} ({url}): {e}")
        # Assuming your function can handle 'None' for a completely failed attempt
        try:
             features = featureExtraction(url, label, None)
        except Exception:
             # If even the fallback fails, append empty/null data so rows align
             features = {"url": url, "error": str(e)}

    legi_features.append(features)

    # Checkpoint: Save to CSV and clear memory every 100 URLs
    if (i + 1) % checkpoint_interval == 0 or (i + 1) == total_urls:
        print(f"Processed {i + 1} URLs. Saving checkpoint...")

        # Check if file exists so we only write headers once
        file_exists = os.path.isfile(csv_filename)

        with open(csv_filename, 'a', newline='', encoding='utf-8') as f:
            # Assuming 'features' is a dictionary. If it's a list, use csv.writer instead.
            if isinstance(features, dict):
                writer = csv.DictWriter(f, fieldnames=features.keys())
                if not file_exists:
                    writer.writeheader()
                writer.writerows(legi_features)
            else:
                writer = csv.writer(f)
                writer.writerows(legi_features)

        # Clear the list to free up RAM!
        legi_features = []

print("Extraction complete!")
```

```python
#converting the list to dataframe
feature_names = ['Domain', 'Have_IP', 'Have_At', 'URL_Length', 'URL_Depth','Redirection',
                      'https_Domain', 'TinyURL', 'Prefix/Suffix', 'DNS_Record', 'Web_Traffic',
                      'Domain_Age', 'Domain_End', 'iFrame', 'Mouse_Over','Right_Click', 'Web_Forwards', 'Label']

# Load both files without headers (as they appear to be raw data)
df1 = pd.read_csv('legi_features_backup (1).csv', header=None, names=feature_names)
df2 = pd.read_csv('legi_features_backup (2).csv', header=None, names=feature_names)
df3 = pd.read_csv('legi_features_backup(3).csv', header=None, names=feature_names)
df4 = pd.read_csv('legi_features_backup(4).csv', header=None, names=feature_names)
df5 = pd.read_csv('legi_features_backup(5).csv', header=None, names=feature_names)

# Concatenate the dataframes
combined_df = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)

combined_df.head()
```
